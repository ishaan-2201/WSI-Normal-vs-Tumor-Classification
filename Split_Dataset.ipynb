```python
source = "/content/Patches"        # contains /normal and /tumor
destination = "/content/data"          # will contain /train, /val, /test
```

```python
def split_dataset(source_dir, dest_dir, split=(0.7, 0.15, 0.15)):
    categories = ['normal', 'tumor']
    for cat in categories:
        src_path = os.path.join(source_dir, cat)
        files = [f for f in os.listdir(src_path) if f.endswith(".png")]
        random.shuffle(files)

        total = len(files)
        train_cut = int(total * split[0])
        val_cut = int(total * (split[0] + split[1]))

        subsets = {
            'train': files[:train_cut],
            'val': files[train_cut:val_cut],
            'test': files[val_cut:]
        }

        for subset, subset_files in subsets.items():
            target_path = os.path.join(dest_dir, subset, cat)
            os.makedirs(target_path, exist_ok=True)
            for f in subset_files:
                shutil.copy(os.path.join(src_path, f), os.path.join(target_path, f))
```

```python
split_dataset(source, destination)
```
